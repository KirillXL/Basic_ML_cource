### 2.1 Какого музыканта Beatles я загадал или entropy reduction

```python
# Появилась новая библиотека - sklearn
from sklearn import tree
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Учебный датасет
data = pd.DataFrame({'X_1': [1, 1, 1, 0, 0, 0, 0, 1], 'X_2': [0, 0, 0, 1, 0, 0, 0, 1], 'Y': [1, 1, 1, 1, 0, 0, 0, 0]})
```

```python
data
```

**Наша задача - обучить предсказывать Y по X_1 и X_2**

```python
clf = tree.DecisionTreeClassifier(criterion = 'entropy')
```

```python
clf #В старых версиях можно было глянуть особенности созданного дерева, сейчас просто класс
```

```python
#Разделяем данные для обучения
X = data[['X_1','X_2']]
y = data.Y
```

```python
#Обучаем
clf.fit(X,y)
```

```python
#Анатолий Карпов не останавливался на визуализации, поэтому я попросил нейросеть создать этот код
# Визуализация данных
plt.figure(figsize=(6, 6))
plt.scatter(data['X_1'][y == 1], data['X_2'][y == 1], c='green', label='Class 1', s=100)
plt.scatter(data['X_1'][y == 0], data['X_2'][y == 0], c='red', label='Class 0', s=100)
plt.xlabel('X_1')
plt.ylabel('X_2')
plt.title('Исходные данные')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#визуализация кода от человека с курса
# X[0] = X_1 , X[1] = X_2
from sklearn.tree import plot_tree
plot_tree(clf,filled=True); 
```
