**Есть такое понятие - пандовская серия** Серия - это строка, которая содержит какое-то число колонок. Отличие от ДатаФрейма в том, что Серия только одномерная( ДатаФрейм двумерный, можно перемещаться по строкам и колонкам) **Личный лайфхак** После выполения команды "из серии" выводятся строки, а при выполении команд над ДатаФреймом - таблица

```python
mean_reading_score = df['reading score'].mean() #Это пандовская серия, мы находим среднее значение
print(mean_reading_score)
```

**Ты можешь запупаться** , мол "ЧЗХ? Если я выведу несколько строк из таблицы - это будет серия, но не таблица? Да, разница в том, что все "серии" не зависят друг от друга и выводятся словно командой принт

### Но иногда нужно объединить фильтры в одно условие и мы привыкли на Питоне писать and или or. НО! С пандовскими сериями все обстоит по другому - у нас условия не выводят True или False, а выводятся ИМЕННО пандовские серии( аля строки) поэтому для нихз используются & или |

```python
(df.gender == "male") & (df['reading score'] > mean_reading_score)
```

```python
query = (df.gender == "male") & (df['reading score'] > mean_reading_score) # сохраняем запрос
```

```python
df.loc[query]
```
